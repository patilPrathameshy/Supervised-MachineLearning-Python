## Data Cleaning is not required as we don't have any numberical column as object

## Label Encoding

```python
# separating numerical and categorical columns
num_cols = df.select_dtypes(["int", "float"])
cat_cols = df.select_dtypes(["object"])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoders = {}
for column in cat_cols:
    le = LabelEncoder()
    cat_cols[column] = le.fit_transform(cat_cols[column])
    label_encoders[column] = le
```

## Scaling numerical columns

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
num_cols_scaled = sc.fit_transform(num_cols)
```

```python
for i in range(len(num_cols.columns)):
    num_cols[num_cols.columns[i]] = num_cols_scaled[:,i]
```

```python
num_cols
```

## Concatinating Scaled numerical cols and label encoded categorical cols

```python
df1 = pd.concat([num_cols,cat_cols], axis=1)
df1
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df1.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
x = df1.iloc[:,:-1]
y = df1.iloc[:,-1]
```

## Data is imbalanced , need to do sampling using Oversampling or Undersampling

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
x_ros, y_ros = ros.fit_resample(x, y)
x=x_ros
y=y_ros
```

## The Data is cleaned and can be used for training

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
# Spliting dataset into train and test data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
knn = KNeighborsClassifier()
```

```python
def mymodel(model):
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(classification_report(y_test,y_pred))
    return model
```

## KNN

```python
mymodel(knn)
```

# Hyperparameter Tuning using GridSearchCV

```python
grid_params = { 'n_neighbors' : [5,7,9,11,13,15],
               'weights' : ['uniform','distance'],
               'metric' : ['minkowski','euclidean','manhattan']}
gs_knn = GridSearchCV(knn, grid_params, verbose = 1, cv=3, n_jobs = 2)
```

```python
gs_knn.fit(x_train.head(10000),y_train.head(10000))
```

```python
gs_knn.best_params_
```

```python
knn_hypertuned = KNeighborsClassifier(metric= 'manhattan', n_neighbors= 15, weights= 'distance')
mymodel(knn_hypertuned)
```

```python
print("K Neighbors Classifer:",knn.score(x_train,y_train))
print("Hypertuned K Neighbors Classifer:", knn_hypertuned.score(x_train,y_train))
```

## From above Output we can say that Hypertuned K Neighbors Classifer is best to use
